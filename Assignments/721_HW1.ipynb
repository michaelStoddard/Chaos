```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
```

```python
x, dx, dt, t, a = smp.symbols('x dx dt t a')

r = smp.symbols('r', positive=True)

f = r*x*(x-a)

f
```

```python
smp.solve(f)
```

```python
def dthe (X1,X2):
    return X2 
def dv (X1,X2, L=1):
    g = -9.8
    return g/L*np.sin(X1)
    
```

```python
# the = np.arange(0,2*np.pi,0.5)
the = np.linspace(np.pi*0 - 0.1,np.pi*0 + 0.1,20)
v = np.linspace(-1,1,20)
```

```python
The, V = np.meshgrid(the,v)
```

```python
U = dthe(The,V)
W = dv(The,V)

M = np.hypot(U,W)
# M = U

plt.contourf(The,V,M)

plt.quiver(The,V,U,W)
# plt.xticks([0,np.pi/2, np.pi, 3*np.pi/2, 2*np.pi],
#            ['0', '$0.5\pi$', '$\pi$', '$1.5\pi$', '$2\pi$'])

plt.xlabel(r'$\theta$')

plt.ylabel(r'$v=\dot \theta$')

plt.show()
```
