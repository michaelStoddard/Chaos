# Van der Pol

```python
x1, x2, a = smp.symbols(r'x_1 x_2 a')

dx1 = x2
dx2 = -x1 + a*x2*(1-x1**2)
```

```python
Jac = smp.Matrix([[smp.diff(dx1,x1),smp.diff(dx1,x2)],
                  [smp.diff(dx2,x1),smp.diff(dx2,x2)]])
Jac
```

```python
J0 = Jac.subs([[x1,0],[x2,0],[a,0]])
J0
```

```python
J0.eigenvects()
```

# Duffing Oscillator

Unforced, undamped

$$\ddot{x} - x + x^3 =0$$  

When linearized

$$\begin{cases}\dot x_1 = x_2\\
\dot x_2 = x_1 - x_1^3 - \delta x_2\end{cases}$$

Modified equation [[ch 4, Wiggins]] 

$$\begin{cases}\dot x_1 = x_2\\
\dot x_2 = x_1 - x_1^3 - \delta x_2 + x_1^2x_2\end{cases}$$


Define symbolic system

```python
x1, x2, dta = smp.symbols(r'x_1 x_2 \delta')

dx1 = x2
dx2 = x1 - x1**3 - dta*x2 + x1**2*x2
```

### Check Zeros

```python
fix_pnts = smp.solve([dx1,dx2],[x1,x2])
fix_pnts
```
